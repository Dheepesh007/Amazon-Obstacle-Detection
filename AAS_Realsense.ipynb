```python
# Obstacle Detection using depth value

# sample depth values (meters)
depth_values = [0.8, 0.6, 0.4, 1.2, 0.3]

threshold = 0.5

for distance in depth_values:

    if distance < threshold:
        velocity = 0
        print("Distance:", distance, "m → Robot Halted")
    else:
        velocity = 1
        print("Distance:", distance, "m → Robot Moving")
```
